```python
import missingno as msno
```

```python
date_soutien=data['Date de soutenance']
date_inscription=data['Date de premiere inscription en doctorat']
date_soutien_nullity=date_soutien.isna()
date_inscription_nullity=date_inscription.isna()
missing_percentage_date_soutien=date_soutien_nullity.mean()*100
missing_percentage_date_inscription=date_inscription_nullity.mean()*100
print (missing_percentage_date_soutien )
print (missing_percentage_date_inscription )

dateData=pd.DataFrame({'Date de soutenance':date_soutien,
                'Date de premiere inscription en doctorat': date_inscription
                })

print(dateData.info())

print(dateData.describe())
print(dateData.head())

dateData.isna().sum()

msno.bar(dateData)
msno.matrix(dateData)
      

```
